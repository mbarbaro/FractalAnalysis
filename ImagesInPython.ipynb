```python
zoom = binbw2[200:400, 0:200] # binbw2 is thresholded to be either black or white
zoom2 = binbw2[590:790, 1150:1350]
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
sbplt = plt.imshow(zoom, cmap = plt.cm.gray)
a.set_title('200-400, 0-200')
a = fig.add_subplot(1, 2, 2)
sbplt = plt.imshow(zoom2, cmap = plt.cm.gray)
a.set_title('590-800, 1150-1350')
```
